```python
# Why list? Compare 5 separate variables vs 1 list variable

# Option A: five separate variables
course1 = 'CS 003A'
course2 = 'CS 003B'
course3 = 'CS 008'
course4 = 'CS 020'
course5 = 'CS 021'

print('Using separate variables:')
print(course1, course2, course3, course4, course5)

# Option B: one list variable
ml_path_courses = ['CS 003A', 'CS 003B', 'CS 008', 'CS 020', 'CS 021']

print('\nUsing one list variable:')
print(ml_path_courses)
print('First course (index 0):', ml_path_courses[0])
print('Total courses:', len(ml_path_courses))
```

## Why introduce lists first?
If you store five courses in five variables, you must manage each name one by one.
A list lets you store all courses in one place, then loop, count, and update easily.

Example: `ml_path_courses` is easier to print, edit, and pass to functions than `course1` ... `course5`.

```python
# List operations on the same course plan
ml_path_courses = ['CS 003A', 'CS 003B', 'CS 008', 'CS 020', 'CS 021']
print('Start:', ml_path_courses)

ml_path_courses.append('CS 025')
ml_path_courses.remove('CS 020')

print('After updates:', ml_path_courses)
```

```python
# Loop through list items
for i, course in enumerate(ml_path_courses, start=1):
    print(f'{i}. {course}')
```

## Part 2: Dictionary basics
A dictionary stores key-value pairs.
Use it when you want lookup by name/key.

```python
weekly_hours = {
    'study': 8,
    'work': 20,
    'commute': 6,
}

print('Study hours:', weekly_hours['study'])
weekly_hours['study'] = 10
weekly_hours['exercise'] = 3

for area, hrs in weekly_hours.items():
    print(f'{area}: {hrs} hours')
```

## Guided Practice 1: Lists (2-3 minutes)
Work with a partner and compare approaches.

Task: You have a list of assignment deadlines.
1. Remove duplicates while keeping first-seen order.
2. Print how many unique deadlines remain.

Data: `deadlines = ['Mon', 'Wed', 'Mon', 'Fri', 'Wed']`

```python
# Starter (students edit this cell)
deadlines = ['Mon', 'Wed', 'Mon', 'Fri', 'Wed']
unique_days = []

for day in deadlines:
    if day not in unique_days:
        unique_days.append(day)

print('Unique days:', unique_days)
print('Count:', len(unique_days))
```

## Integrated example: Weekly task completion
Combine a list of tasks with a dictionary of completion counts.

```python
tasks = ['reading', 'coding practice', 'reading', 'quiz prep', 'coding practice']
completed = {}

for task in tasks:
    completed[task] = completed.get(task, 0) + 1

print('Task list:', tasks)
print('Completion counts:', completed)
```

```python
for task, count in completed.items():
    print(f'{task}: {count}')

most_repeated = max(completed, key=completed.get)
print('Most repeated task:', most_repeated)
```

## Beginner tips to remember
- Use a list when order matters.
- Use a dictionary when lookup by name/key matters.
- Lists are indexed starting at 0.
- `.append()` adds to a list.
- `dict.get(key, default)` avoids key errors.

```python
sample = ['a', 'b', 'c']
print('index 0 ->', sample[0])
print('index 1 ->', sample[1])
print('index 2 ->', sample[2])
```
