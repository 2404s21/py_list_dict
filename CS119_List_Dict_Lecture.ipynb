## Compare the structures (quick mental model)

```python
from dataclasses import dataclass

playlist = ['Lo-fi', 'Study Beats', '8-bit Remix']              # list: ordered + mutable
office_hours = ('Mon 2pm', 'Wed 2pm')                           # tuple: ordered + immutable
student = {'name': 'Ana', 'major': 'CS', 'units': 15}           # dict: key -> value mapping

@dataclass
class Course:
    code: str
    room: str

course = Course('CS119', 'CHEM 101')                            # object: behavior + state

print('list   :', playlist)
print('tuple  :', office_hours)
print('dict   :', student)
print('object :', course)
```

```python
# Important spec idea: mutability
playlist.append('Synthwave')
student['units'] += 1
print('Mutated list:', playlist)
print('Mutated dict:', student)

try:
    office_hours[0] = 'Fri 2pm'
except TypeError as e:
    print('Tuple is immutable ->', e.__class__.__name__)
```

## Demo A: Boba budget tracker (student life + funny)

```python
orders = [
    ('Ana', 'Brown Sugar', 6.50),
    ('Ben', 'Taro', 5.75),
    ('Ana', 'Matcha', 6.25),
    ('Cory', 'Thai Tea', 5.50),
    ('Ben', 'Brown Sugar', 6.50),
]

spend = {}
for name, drink, price in orders:
    spend[name] = spend.get(name, 0) + price

print('Total spend by student:')
for name, total in spend.items():
    print(f'{name:4} -> ${total:.2f}')
```

```python
# Important spec idea: dict preserves insertion order (Python 3.7+)
print(list(spend.items()))
```

## Demo B: Meme reaction leaderboard
Uses list iteration + dict counting

```python
reactions = ['LOL', 'LOL', 'OMG', 'LOL', 'nice', 'OMG', 'nice', 'nice', 'nice']
leaderboard = {}

for r in reactions:
    key = r.lower()
    leaderboard[key] = leaderboard.get(key, 0) + 1

for reaction, count in sorted(leaderboard.items(), key=lambda kv: kv[1], reverse=True):
    print(f'{reaction:5} -> {count}')
```

## In-class Exercise 2 (2 minutes)
Build a profile object and test with your own data.

Task: Use a class to model a student profile.
Include:
- name (str)
- clubs (list[str])
- gpa (float)
Then print a one-line intro.

```python
from dataclasses import dataclass

@dataclass
class StudentProfile:
    name: str
    clubs: list
    gpa: float

    def intro(self):
        return f"{self.name} | clubs={len(self.clubs)} | GPA={self.gpa:.2f}"

p = StudentProfile('Dina', ['Robotics', 'Game Dev Club'], 3.72)
print(p.intro())
```

## Integrated Example: Campus event check-in system
Combines tuple, list, dict, and object.

```python
from dataclasses import dataclass

@dataclass
class Event:
    name: str
    room: str

event = Event('Hack Night', 'SCI-201')
sessions = ('5:00 PM', '6:00 PM', '7:00 PM')          # tuple: schedule should not change
queue = ['Ana', 'Ben', 'Ana', 'Cory', 'Dina', 'Ben'] # list: arrival order can repeat
checkins = {}                                         # dict: name -> check-in count

for name in queue:
    checkins[name] = checkins.get(name, 0) + 1

print(f'Event: {event.name} in {event.room}')
print('Sessions:', sessions)
print('Check-ins:', checkins)
```

```python
# Typical and important: comprehensions + unpacking
multi_checkin_students = [name for name, cnt in checkins.items() if cnt > 1]
first, *middle, last = queue
print('Multiple check-ins:', multi_checkin_students)
print('First arrival:', first, '| Last arrival:', last)
```

## Python language specs worth remembering
- list and dict are mutable
- tuple is immutable
- dict keys must be hashable (e.g., str, int, tuple of immutables)
- dict preserves insertion order (Python 3.7+)
- object identity (`is`) is different from value equality (`==`)

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a
print('a == b:', a == b)
print('a is b:', a is b)
print('a is c:', a is c)
```
