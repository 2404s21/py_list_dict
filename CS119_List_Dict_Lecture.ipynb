## Lists: quick ideaA list is an ordered collection. It can change (mutable).

```python
courses = ['CS 003A', 'MATH 227', 'ENGLISH 101']print('First course:', courses[0])print('Total courses:', len(courses))
```

### Lists can hold different typesThis is valid Python (even if we usually keep one type for clarity).

```python
mixed = ['CS 003A', 4, True, 3.7]print(mixed)print('Type of 2nd item:', type(mixed[1]).__name__)
```

## Dictionaries: quick ideaA dictionary maps a key to a value.Use it when you need fast lookup by name.

```python
student = {'name': 'Alex', 'major': 'CS', 'units': 12}print(student['name'])student['units'] = 15print(student)
```

### Example: median salaries (practice data)Practice with dictionary lookups using approximate sample values.

```python
median_salary = {'Data Analyst': 78000, 'ML Engineer': 145000, 'Software Engineer': 130000}print('ML Engineer median:', median_salary['ML Engineer'])
```

### Example: transfer requirements (practice structure)Store requirements by university in a dictionary of dictionaries.

```python
transfer = {'UCLA': {'calc': 1, 'python': 1}, 'CSUN': {'calc': 1, 'python': 0}, 'UCI': {'calc': 2, 'python': 1}}for school, req in transfer.items():    print(school, '-> Python required:', bool(req['python']))
```

## CS uses of lists and dictionariesLists are common for:- queues, stacks, ordered logs, mini datasetsDictionaries are common for:- configuration settings- user profiles by ID- frequency counts- API/JSON data

## This week's contribution points (0.1 to 0.5)Track tasks with a list, and point values with a dictionary.

```python
tasks = ['go to office hours', 'ask 5 academic questions', 'share one resource']points = {'go to office hours': 0.5, 'ask 5 academic questions': 0.3, 'share one resource': 0.2}total = 0.0for t in tasks:    total += points[t]print('Total contribution points:', round(total, 2))
```

## Tuple and object (very brief)- tuple: like a list, but immutable (cannot change)- object: groups data + behavior in one type

```python
from dataclasses import dataclasstime_slot = ('Mon', '2:00 PM')@dataclassclass StudentCard:    name: strprint(time_slot, StudentCard('Maya'))
```
