## 1. Lists
- `list`: an ordered, changeable collection of items.
- Lists use index positions starting at `0`.

Use a list when:
- order matters (e.g. course sequence, checklist)
- you need add/remove/update by position

But first, ***why?***

```python
# Why list? Compare 5 separate variables vs 1 list variable

# Option A: five separate variables (harder to manage as data grows)
course1 = 'CS 003A'
course2 = 'CS 003B'
course3 = 'CS 008'
course4 = 'CS 020'
course5 = 'CS 021'

print('Using separate variables:')
print(course1, course2, course3, course4, course5)

# Option B: one list variable (easier to loop, count, and update)
my_course_plan = ['CS 003A', 'CS 003B', 'CS 008', 'CS 020', 'CS 021']

print('\nUsing one list variable:')
print(my_course_plan)
print('First course (index 0):', my_course_plan[0])
print('Total courses:', len(my_course_plan))
print('Last course:', my_course_plan[-1])
```

## Lists roadmap mapping
- Ordered by position: a list keeps sequence (`0, 1, 2, ...`).
- Accessed by index: use `my_list[0]`, `my_list[-1]`.
- Mutable: we can add, change, and remove items.
- Useful for ordered collections: course plans, checklists, schedules.

The next two code cells show all of these in order.

```python
# List operations: add, change, remove
print('Start:', my_course_plan)

# Add one new course at the end
my_course_plan.append('CS 025')

# Change one item by index
my_course_plan[1] = 'CS 007'

# Remove one course by value
my_course_plan.remove('CS 020')

print('After add/change/remove:', my_course_plan)
```

```python
# Loop through list items
for i, course in enumerate(my_course_plan, start=1):
    print(f'{i}. {course}')
```

## 2. Dictionaries
- Store key-value pairs: example `{'study': 8}`
- Accessed by key: `weekly_hours['study']`
- Mutable: we can add, change, and remove key-value pairs
- Useful for labeled information: hours by category, profile fields, settings

The next code cell follows this same order.

## Common mistakes (quick check)
- List index starts at `0`, not `1`.
- `append()` adds one item to the end; it does not replace existing items.
- Dictionary keys are case-sensitive (`'Name'` is not `'name'`).
- Accessing a missing dictionary key causes an error; use `dict.get(key, default)` when unsure.
- Watch indentation inside `for` loops and `if` blocks.

```python
# Dictionary operations: add, change, remove
weekly_hours = {
    'study': 8,
    'work': 20,
    'commute': 6,
}

# Access by key
print('Study hours:', weekly_hours['study'])

# Change existing value
weekly_hours['study'] = 10

# Add new key-value pair
weekly_hours['exercise'] = 3

# Remove key-value pair
removed = weekly_hours.pop('commute')
print('Removed commute hours:', removed)

# Loop through all key-value pairs
for area, hrs in weekly_hours.items():
    print(f'{area}: {hrs} hours')
```

## Integrated example: Weekly task completion
Combine a list of tasks with a dictionary of completion counts.

```python
tasks = ['reading', 'coding practice', 'reading', 'quiz prep', 'coding practice']
completed = {}

for task in tasks:
    completed[task] = completed.get(task, 0) + 1

print('Task list:', tasks)
print('Completion counts:', completed)
```

```python
for task, count in completed.items():
    print(f'{task}: {count}')

most_repeated = max(completed, key=completed.get)
print('Most repeated task:', most_repeated)
```

## Beginner tips to remember
- `list` = ordered items by position (index starts at 0).
- `dictionary` = key:value pairs by name.
- Use `len(...)` to count items.
- Use `.append(...)` to add to a list.
- Use `dict.get(key, default)` when a key might be missing.
